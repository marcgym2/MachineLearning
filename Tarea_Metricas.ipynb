```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Cargar los datos
from google.colab import drive
drive.mount("/content/drive")
df_measures = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/measures_v2.csv")
```

```python
# Seleccionar las variables independientes (x_vars) y la variable dependiente (y_vars)
x_vars = ["u_q", "coolant", "ambient", "u_d", "motor_speed", "i_q", "i_d"]
y_vars = ["pm", "torque", "stator_yoke", "stator_tooth", "stator_winding"]
```

```python
# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
train = df_measures.sample(frac=0.8, random_state=123)
test = df_measures.drop(train.index)
```

```python
# Definir el modelo
model = Sequential()
model.add(Dense(32, input_dim=len(x_vars), activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(len(y_vars), activation='linear'))
```

```python
# Compilar el modelo
model.compile(loss='mse', optimizer='adam')
```

```python
# Entrenar el modelo
model.fit(train[x_vars], train[y_vars], epochs=100, batch_size=32, verbose=1)
```

```python
# Calcular las predicciones del modelo en el conjunto de prueba
predictions = model.predict(test[x_vars])
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_squared_error

# Calcular las métricas RMSE, R^2 y MAE
rmse = np.sqrt(mean_squared_error(test[y_vars], predictions))
r2 = r2_score(test[y_vars], predictions)
mae = mean_absolute_error(test[y_vars], predictions)
mse = mean_squared_error(test[y_vars], predictions)

# Mostrar las métricas

print('R^2 en el conjunto de prueba:', r2)
print('MAE en el conjunto de prueba:', mae)
print('RMSE en el conjunto de prueba:', rmse)
print('MSE en el conjunto de prueba:', mse)
```
